```python
x = 90
if x > 80:
  print('축하합니다~ 합격입니다')
  print('다음주부터 출근하시면 됩니다.')
elif x > 60:
  print('2차 면접시험 대상자입니다.')
  print('시험 날짜를 알려드리겠습니다.')
else:
  print('불합격입니다.')
```

```python
score = [50, 70, 90]
for x in score:
  print("-----------------------------------")
  if x > 80:
    print('축하합니다~ 합격입니다')
    print('다음주부터 출근하시면 됩니다.')
  elif x > 60:
    print('2차 면접시험 대상자입니다.')
    print('시험 날짜를 알려드리겠습니다.')
  else:
    print('불합격입니다.')
```

```python
score = [50, 70, 90]
for i, x in enumerate(score):
  print("-----------------------------------")
  print(i+1,'번째 참가자님,')
  if x > 80:
    print('축하합니다~ 합격입니다')
    print('다음주부터 출근하시면 됩니다.')
  elif x > 60:
    print('2차 면접시험 대상자입니다.')
    print('시험 날짜를 알려드리겠습니다.')
  else:
    print('불합격입니다.')
```
